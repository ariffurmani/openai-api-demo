```python
import textwrap
# Loading the keys

import os
from dotenv import load_dotenv

import textwrap

def prettyPrint(*args):
    text = " ".join(str(arg) for arg in args)
    try:
        print(textwrap.fill(text, width=80))
    except Exception as e:
        print(text)

# Load env file from the mentioned directory
load_dotenv('/home/furmani-arif/Documents/global-properties.env')

#fetch the openai api key fom the loaded env file
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError(
        "OPENAI_API_KEY not found! "
        "Make sure you have a .env file with OPENAI_API_KEY"
    )

prettyPrint("API key loaded successfully")

```

```python
# Load the OpenAI client
from openai import OpenAI

client = OpenAI(api_key = api_key)
prettyPrint("OpenAI client is ready")
```

```python
# Simple Demo-1
resp1 = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {"role": "developer", "content": "You are very polite and kind assistant"},
        {"role": "user", "content": "My name is Arif"}
    ],
    # instructions="You are sarcastic and a jerk, but still provide with the answer in the end",
)

prettyPrint("Response API Output:", resp1.output_text)
```

```python

resp2 = client.responses.create(
    model="gpt-4o-mini",
    input="What is my name?"
)

prettyPrint("Response API Output:", resp2.output_text)
```

```python
# Uses previous_response_id to remember context and previous question
resp3 = client.responses.create(
    model="gpt-4o-mini",
    previous_response_id=resp1.id,
    input="What is my name?"
)

prettyPrint("Response API Output:", resp3.output_text)
```

```python
# Uses instructions as system prompt
resp = client.responses.create(
    model="gpt-4o-mini",
    previous_response_id=resp1.id,
    input="What is my name?",
    instructions="You are sarcastic and a jerk, but still provide with the answer in the end"
)

prettyPrint("Response API Output:", resp.output_text)
```

```python
resp = client.responses.create(
    model = "gpt-5-nano",
    instructions = "You are a senior profession in the CSE branch of IIT",
    input = "Will AI take our jobs",
    # max_output_tokens=1000,
    reasoning={"effort": "high"}
)

prettyPrint("Response API Output:", resp.output_text, resp.status, resp.incomplete_details)
```
