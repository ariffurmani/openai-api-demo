```python
import textwrap
# Loading the keys

import os
from dotenv import load_dotenv

import textwrap

def prettyPrint(*args):
    text = " ".join(str(arg) for arg in args)
    try:
        print(textwrap.fill(text, width=80))
    except Exception as e:
        print(text)

# Load env file from the mentioned directory
load_dotenv('/home/furmani-arif/Documents/global-properties.env')

#fetch the openai api key fom the loaded env file
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError(
        "OPENAI_API_KEY not found! "
        "Make sure you have a .env file with OPENAI_API_KEY"
    )

prettyPrint("API key loaded successfully")

```

```python
# Load the OpenAI client
from openai import OpenAI

client = OpenAI(api_key = api_key)
prettyPrint("OpenAI client is ready")
```

```python
# Check all the models available
models = client.models.list()

for m in models.data:
    prettyPrint(m)
```

```python
# First attempt with OpenAI's Chat Completions API with role as user
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages=[
        {"role": "user", "content": "What is the capital of India"}
    ]
)

prettyPrint(response.choices[0].message.content)
```

```python
# Try with system prompt as well with role as system
response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are complete jerk, be as unhelpful as possible. But then in the end answer the question"},
        {"role": "user", "content": "What is the capital of India"}
    ]
)

prettyPrint(response.choices[0].message.content)
```

```python
# Play around

current_message = [
    {"role": "system", "content": "You are sarcastic and unhelpful person, but in the end answer the question"},
    {"role": "user", "content": "What is capital of India"}
]

response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = current_message
)

reply = response.choices[0].message.content

prettyPrint(reply)
```

```python
# Check if it remembers the previous conversation

ask_message = [
    {"role": "user", "content": "What was my previous question?"}
]

response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = ask_message
)

reply2 = response.choices[0].message.content

prettyPrint(reply2)
```

```python
# Using role assistant to append previous question as context

current_message.append({"role": "assistant", "content": reply},)
current_message.append({"role": "user", "content": "What was my previous question?"})

response = client.chat.completions.create(
    model = "gpt-4o-mini",
    messages = current_message
)

reply = response.choices[0].message.content

prettyPrint(reply)
```
